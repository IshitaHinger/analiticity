```python
df['income']=df['AMOUNT']/df['RATIO']

s=df[df['income']>0].groupby(['OCC']).mean()['income']

df.loc[(df['income']<0) & (df['OCC']==0.0), 'income'] = s[0.0]
df.loc[(df['income']<0) & (df['OCC']==1.0), 'income'] = s[1.0]

df.loc[(df['income']<0) & (df['OCC']==2.0), 'income'] = s[2.0]

df.loc[(df['income']<0) & (df['OCC']==3.0), 'income'] = s[3.0]
df.loc[(df['income']<0) & (df['OCC']==4.0), 'income'] = s[4.0]

df.loc[(df['income']<0) & (df['OCC']==5.0), 'income'] = s[5.0]

df['RATIO']=df['AMOUNT']/df['income']
```

```python
import pandas as pd
import csv
dt= pd.read_csv("/resources/data/test.csv")
dt.isnull().sum()


print(dt.corr().loc['DUE_MORTGAGE', 'VALUE'])


print(dt.loc[:,['DUE_MORTGAGE', 'VALUE']].head())

#value and DUE_MORTGAGE seem to be in a ratio

dt_1=(dt.loc[:,['DUE_MORTGAGE', 'VALUE']]).dropna()
dt_1['ratio'] = dt_1['VALUE'] / dt_1['DUE_MORTGAGE']

print(dt_1['ratio'].describe())

#checking for large ratios (outliers)

print(dt_1[dt_1['ratio'] > 3].shape)

#excluding these outliers and taking mean of ratio
V_D = dt_1[dt_1['ratio'] < 3].loc[:,'ratio'].mean()

print(V_D)

#use V_D for imputing DUE_MORTGAGE and VALUE as required



#finding rows where both DUE_MORTGAGE and VALUE are NaN
## print(df[df['DUE_MORTGAGE'].isnull() & df['VALUE'].isnull()].shape)

#it is a small no; imputing all such rows with means and other rows with the multiplied ratio
dt.loc[dt['DUE_MORTGAGE'].isnull() & dt['VALUE'].isnull(), 'DUE_MORTGAGE'] = dt['DUE_MORTGAGE'].mean()
dt.loc[dt['DUE_MORTGAGE'].isnull() & dt['VALUE'].isnull(), 'VALUE'] = dt['VALUE'].mean()

dt['DUE_MORTGAGE'].fillna(dt['VALUE']/V_D, inplace=True)
dt['VALUE'].fillna(dt['DUE_MORTGAGE']*V_D, inplace=True)



dt.isnull().sum()
##print(df['REASON'].describe())
dt['REASON'].fillna(dt['REASON'].mode()[0], inplace=True)
dt['OCC'].fillna(dt['OCC'].mode()[0], inplace=True)
dt['DCL'].fillna(dt['DCL'].mode()[0], inplace=True)
dt['CLT'].fillna(dt['CLT'].mean(), inplace=True)
dt['TJOB'].fillna(dt['TJOB'].mean(), inplace=True)  #doubtful
dt['CL_COUNT'].fillna(dt['CL_COUNT'].mean(), inplace=True)
dt['RATIO'].fillna(-1, inplace=True)  #replace with -1 while modeling

dt.isnull().sum()

#income
dt['income']=dt['AMOUNT']/dt['RATIO']

#imputing values of ratio
s=dt[dt['income']>0].groupby(['OCC']).mean()['income']

dt.loc[(dt['income']<0) & (dt['OCC']==0.0), 'income'] = s[0.0]
dt.loc[(dt['income']<0) & (dt['OCC']==1.0), 'income'] = s[1.0]

dt.loc[(dt['income']<0) & (dt['OCC']==2.0), 'income'] = s[2.0]

dt.loc[(dt['income']<0) & (dt['OCC']==3.0), 'income'] = s[3.0]
dt.loc[(dt['income']<0) & (dt['OCC']==4.0), 'income'] = s[4.0]

dt.loc[(dt['income']<0) & (dt['OCC']==5.0), 'income'] = s[5.0]

dt['RATIO']=dt['AMOUNT']/dt['income']

```
